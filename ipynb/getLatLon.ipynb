```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL of page to be scraped
url = 'https://developers.google.com/public-data/docs/canonical/countries_csv'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
table = soup.find('table')
```

```python
header = table.find_all('th')
rows = table.find_all('tr')
```

```python
column_names = [th.text for th in header]
column_names
```

```python
# Loop through returned results
l = []
for tr in rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)
    
countries_df = pd.DataFrame(l, columns = column_names)
```

```python
countries_df = countries_df.set_index('name')
```

```python
countries_df.dropna(inplace=True)
```

```python
countries_df = countries_df.sort_index()
```

```python
countries_df.to_csv("static/csv/countries.csv")
```
